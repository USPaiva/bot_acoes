```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re


def get_detalhes_acao_oceans14(nome, ticker):
    url = f"https://www.oceans14.com.br/acoes/{nome.lower()}/{ticker.lower()}/balanco-dividendos"
    #headers = {"User-Agent": "Mozilla/5.0"}
    resp = requests.get(url, headers=headers)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")

    def get_by_id(id_):
        el = soup.find(id=id_)
        return el.get_text(strip=True) if el else None

    def buscar_variacao_12m():
        el = soup.find(id="ctl00_conteudoPrincipal_lblCotacaoVariacao12mTicker01Desktop")
        if el:
            match = re.search(r"Var\. 12m:\s*(.*)", el.text)
            return match.group(1).strip() if match else None
        return None

    def get_lista_abas():
        abas = soup.select(".nav-tabs > li > span")
        return [aba.get_text(strip=True) for aba in abas if aba.get("data-target")]

    def get_links_graficos():
        links = []
        for aba in ["abaCotacao", "abaCotacaoLucro", "abaHistoricoPL"]:
            if soup.find(id=aba):
                links.append(f"https://www.oceans14.com.br/acoes/{nome.lower()}/{ticker.lower()}/{aba}")
        return links

    def extrair_valor_por_titulo(titulo):
        # Procura o título nas divs do painel financeiro
        linhas = soup.select("#abaResultados .row")
        for linha in linhas:
            colunas = linha.find_all("div")
            if len(colunas) >= 2:
                nome = colunas[0].text.strip().lower()
                valor = colunas[1].text.strip()
                if titulo.lower() in nome:
                    return valor
        return None

    # Extração básica
    dados = {
        "nome_url": nome.upper(),
        "ticker": ticker.upper(),
        "razao_social": get_by_id("ctl00_conteudoPrincipal_lblRazaoSocial"),
        "nome_fantasia": soup.select_one("span[style*='font-size:16px']").text.strip() if soup.select_one("span[style*='font-size:16px']") else None,
        "cnpj": get_by_id("ctl00_conteudoPrincipal_lblCnpj"),
        "quantidade_acoes": get_by_id("ctl00_conteudoPrincipal_lblNumeroAcoes"),
        "free_float": get_by_id("ctl00_conteudoPrincipal_lblFreeFloat"),
        "quantidade_acionistas": get_by_id("ctl00_conteudoPrincipal_lblQntAcionistas"),
        "valor_mercado": get_by_id("ctl00_conteudoPrincipal_lblValorMercado"),
        "cotacao_atual": get_by_id("ctl00_conteudoPrincipal_lblCotacaoTicker01Desktop"),
        "variacao_dia": soup.select_one("#ctl00_conteudoPrincipal_lblCotacaoVariacao1dTicker01Desktop div").text.strip() if soup.select_one("#ctl00_conteudoPrincipal_lblCotacaoVariacao1dTicker01Desktop div") else None,
        "variacao_12m": buscar_variacao_12m(),
        "data_hora_cotacao": get_by_id("ctl00_conteudoPrincipal_lblCotacaoDataHoraTicker01Desktop"),
        "segmento_listagem": get_by_id("ctl00_conteudoPrincipal_lblGovernanca"),
        "setor": get_by_id("ctl00_conteudoPrincipal_lblSetor"),
        "subsetor": get_by_id("ctl00_conteudoPrincipal_lblSubsetor"),
        "segmento": get_by_id("ctl00_conteudoPrincipal_lblSegmento"),
        "site_ri": get_by_id("ctl00_conteudoPrincipal_lblSiteRI"),
        "ultimo_resultado": get_by_id("ctl00_conteudoPrincipal_lblUltimoBalanco"),
        "proximo_resultado": get_by_id("ctl00_conteudoPrincipal_lblProximoBalanco"),
        "abas_disponiveis": get_lista_abas(),
        "links_graficos": get_links_graficos(),
    }

    # Financeiros (demonstrativo)
    receita_liquida = extrair_valor_por_titulo("receita líquida")
    lucro_bruto = extrair_valor_por_titulo("lucro bruto")
    lucro_liquido = extrair_valor_por_titulo("lucro líquido")
    ebitda = extrair_valor_por_titulo("ebitda")
    ebit = extrair_valor_por_titulo("ebit")

    dados["receita_liquida"] = receita_liquida
    dados["lucro_bruto"] = lucro_bruto
    dados["lucro_liquido"] = lucro_liquido
    dados["ebitda"] = ebitda
    dados["ebit"] = ebit

    # Cálculo de margem bruta se possível
    try:
        receita_valor = float(receita_liquida.replace(".", "").replace(",", ".").replace("R$", "").replace("mil", "").strip())
        lucro_valor = float(lucro_bruto.replace(".", "").replace(",", ".").replace("R$", "").replace("mil", "").strip())
        margem_bruta = (lucro_valor / receita_valor) * 100 if receita_valor != 0 else None
        dados["margem_bruta_percentual"] = f"{margem_bruta:.2f}%"
    except:
        dados["margem_bruta_percentual"] = None

    return pd.DataFrame([dados])


# Exemplo de uso
if __name__ == "__main__":
    df = get_detalhes_acao_oceans14("weg", "WEGE3")
    pd.set_option("display.max_rows", None)
    print(df.T)
```

# Estrategia Bazin

```python
"""
Metodo Bazin - seleção e backtest simples
Dependências: yfinance, pandas, numpy, requests
pip install yfinance pandas numpy requests
"""

import yfinance as yf
import pandas as pd
import numpy as np
import requests
from datetime import datetime, timedelta
from time import sleep

# ---------- Configurações ----------
TICKERS = [
    # exemplo: coloque aqui os tickers (yfinance symbol). Para B3 usar sufixo '.SA' ex: 'VALE3.SA'
    "VALE3.SA", "ITUB4.SA", "PETR4.SA", "BBDC4.SA", "BBAS3.SA"
]
AS_OF = datetime.today().date()            # data de avaliação (ponto no tempo)
LOOKBACK_DAYS = 365                        # trailing 12 meses para dividendos
MIN_CASH_YIELD = 0.06                      # 6% a.a. (Bazin)
MIN_LIQUIDITY_BRL = 200_000               # min valor médio diário negociado em BRL (ajuste conforme desejar)
MAX_DEBT_TO_EBITDA = 4.0                   # opcional, None para não aplicar
TOP_N = 10                                 # número de ações a selecionar
DOLLARIZE = True                           # True para converter proventos/preço para USD
DOLLARIZE_MODE = "historical"              # "latest" ou "historical"
# ------------------------------------

# Cache para taxas FX
FX_CACHE = {}

def get_fx_rate_br_to_usd(date: datetime.date, mode="latest"):
    """
    Retorna taxa BRL->USD (1 BRL = x USD) usando exchangerate.host (gratuito).
    mode: "latest" usa taxa atual; "historical" usa a taxa na data.
    Retorna None se falhar.
    """
    if not DOLLARIZE:
        return 1.0
    key = (date.isoformat(), mode)
    if key in FX_CACHE:
        return FX_CACHE[key]
    try:
        if mode == "latest":
            url = "https://api.exchangerate.host/latest?base=BRL&symbols=USD"
        else:
            url = f"https://api.exchangerate.host/{date.isoformat()}?base=BRL&symbols=USD"
        r = requests.get(url, timeout=10)
        r.raise_for_status()
        data = r.json()
        rate = data["rates"]["USD"]  # 1 BRL = rate USD
        FX_CACHE[key] = rate
        sleep(0.1)  # evitar throttling
        return rate
    except Exception:
        return None

def fetch_dividends_and_prices(ticker, start_date, end_date):
    """
    Retorna DataFrame de preços diários (adj close, volume) e Series de dividendos (date -> value)
    """
    tk = yf.Ticker(ticker)
    # histórico de preços
    hist = tk.history(start=start_date - timedelta(days=5), end=end_date + timedelta(days=1), actions=False, auto_adjust=False)
    if hist.empty:
        return None, None
    hist = hist.rename(columns={"Close": "close", "Adj Close": "adj_close", "Volume": "volume"})
    # dividendos
    divs = tk.dividends
    if isinstance(divs, pd.Series):
        divs = divs[ (divs.index.date >= start_date) & (divs.index.date <= end_date) ]
    else:
        divs = pd.Series(dtype=float)
    return hist, divs

def trailing_12m_dividend_sum(divs, as_of_date):
    """
    Soma dos dividendos pagos nos 12 meses anteriores a as_of_date (exclui pagamentos futuros).
    divs: pd.Series indexed por Timestamp
    """
    start = as_of_date - timedelta(days=365)
    mask = (divs.index.date > start) & (divs.index.date <= as_of_date)
    return float(divs.loc[mask].sum()) if not divs.empty else 0.0

def average_daily_traded_value_brl(hist):
    """
    Estima liquidez como média (volume * close) diária nos últimos 252 pregões do DataFrame hist.
    hist deve conter colunas close e volume, index datas.
    """
    df = hist.copy()
    df = df.dropna(subset=["close", "volume"])
    df["traded_value"] = df["close"] * df["volume"]
    # usar últimos 252 pregões se disponíveis
    last = df.tail(252)
    return float(last["traded_value"].mean()) if not last.empty else 0.0

def get_debt_to_ebitda(ticker):
    """
    Tenta obter totalDebt / ebitda via yfinance info.
    Pode retornar None se dados ausentes.
    """
    tk = yf.Ticker(ticker)
    info = tk.info
    try:
        total_debt = info.get("totalDebt")  # em moeda local
        ebitda = info.get("ebitda")
        if total_debt is None or ebitda is None or ebitda == 0:
            return None
        return total_debt / abs(ebitda)
    except Exception:
        return None

def evaluate_ticker(ticker, as_of_date):
    """
    Avalia um ticker com os critérios do Bazin.
    Retorna dicionário com métricas (cash_yield, dividend_yield, liquidity, debt_ratio, passes_filters)
    """
    end = as_of_date
    start = as_of_date - timedelta(days=LOOKBACK_DAYS + 30)
    hist, divs = fetch_dividends_and_prices(ticker, start.date(), end.date())
    if hist is None or hist.empty:
        return {"ticker": ticker, "error": "no price data"}
    # preço atual -> último close
    last_price = float(hist["close"].dropna().iloc[-1])
    # trailing divs
    trailing_divs = trailing_12m_dividend_sum(divs, as_of_date)
    # Dividend Yield (local)
    dividend_yield = trailing_divs / last_price if last_price > 0 else 0.0

    # Dolarizar (opcional)
    if DOLLARIZE:
        if DOLLARIZE_MODE == "historical":
            # converter cada pagamento para USD na data e somar; converter preço na as_of_date
            cash_usd = 0.0
            for dt, val in divs.items():
                if dt.date() <= as_of_date and dt.date() > (as_of_date - timedelta(days=365)):
                    rate = get_fx_rate_br_to_usd(dt.date(), mode="historical")
                    if rate is None:
                        rate = get_fx_rate_br_to_usd(as_of_date, mode="latest")
                    cash_usd += float(val) * (rate if rate is not None else 1.0)
            # preço em USD usando rate on as_of_date
            rate_price = get_fx_rate_br_to_usd(as_of_date, mode="historical")
            if rate_price is None:
                rate_price = get_fx_rate_br_to_usd(as_of_date, mode="latest")
            price_usd = last_price * (rate_price if rate_price is not None else 1.0)
            cash_yield = (cash_usd / price_usd) if price_usd > 0 else 0.0
        else:
            # usar taxa atual (aprox)
            rate = get_fx_rate_br_to_usd(as_of_date, mode="latest") or 1.0
            cash_usd = trailing_divs * rate
            price_usd = last_price * rate
            cash_yield = (cash_usd / price_usd) if price_usd > 0 else 0.0
    else:
        cash_yield = dividend_yield

    # Liquidez
    liquidity_brl = average_daily_traded_value_brl(hist)

    # Endividamento (opcional)
    debt_ratio = get_debt_to_ebitda(ticker)

    # Aplicar filtros
    passes = True
    reasons = []
    if cash_yield < MIN_CASH_YIELD:
        passes = False
        reasons.append("cash_yield_below_min")
    if liquidity_brl < MIN_LIQUIDITY_BRL:
        passes = False
        reasons.append("low_liquidity")
    if MAX_DEBT_TO_EBITDA is not None and debt_ratio is not None and debt_ratio > MAX_DEBT_TO_EBITDA:
        passes = False
        reasons.append("high_debt_to_ebitda")

    return {
        "ticker": ticker,
        "last_price": last_price,
        "trailing_divs": trailing_divs,
        "dividend_yield": dividend_yield,
        "cash_yield": cash_yield,
        "liquidity_brl": liquidity_brl,
        "debt_to_ebitda": debt_ratio,
        "passes": passes,
        "reasons": reasons
    }

def select_universe(tickers, as_of_date):
    results = []
    for t in tickers:
        print(f"avaliando {t} ...")
        try:
            r = evaluate_ticker(t, as_of_date)
            results.append(r)
        except Exception as e:
            results.append({"ticker": t, "error": str(e)})
        sleep(0.1)
    df = pd.DataFrame(results)
    # filtrar apenas que passaram
    df_pass = df[df["passes"] == True].copy()
    df_pass = df_pass.sort_values("cash_yield", ascending=False)
    selection = df_pass.head(TOP_N)
    return df, selection

# Função de backtest simples (equal-weight, reinveste dividendos em caixa e rebalanceia)
def simple_backtest(selection_tickers, start_date, end_date, initial_capital=10000.0, rebalance_months=6):
    """
    Backtest simples:
    - equal weight at each rebalance date
    - dividend payments adicionados ao caixa e reinvestidos somente no rebalance
    - preços usados: close
    """
    # calendário de rebalance
    dates = pd.date_range(start=start_date, end=end_date, freq='M').to_pydatetime().tolist()
    # filtrar rebalance months
    rebalance_dates = [d.date() for i,d in enumerate(dates) if (i % rebalance_months) == 0]
    # init portfolio
    cash = initial_capital
    holdings = {t: 0 for t in selection_tickers}
    last_prices = {}
    # dataframe para resultados
    rows = []
    for dt in rebalance_dates:
        # fetch prices at dt
        prices = {}
        for t in selection_tickers:
            tk = yf.Ticker(t)
            # pegar preço mais próximo anterior ao dt
            hist = tk.history(end=dt + timedelta(days=1), period="60d")
            if hist.empty:
                prices[t] = None
            else:
                prices[t] = float(hist["Close"].dropna().iloc[-1])
            sleep(0.05)
        # reinvest cash + dividendos since last rebalance
        # (simple approach: get dividends between last rebalance and dt)
        if sum(1 for p in prices.values() if p is None) > 0:
            # se dados faltantes, pular este rebalance
            continue
        # compute total portfolio value (holdings * prices + cash)
        if any(p is None for p in prices.values()):
            continue
        value_holdings = sum(holdings[t] * prices[t] for t in selection_tickers)
        total_value = value_holdings + cash
        # equal weight allocation
        alloc_per_ticker = total_value / len(selection_tickers)
        # rebalance: compute new holdings
        for t in selection_tickers:
            holdings[t] = alloc_per_ticker / prices[t]
            last_prices[t] = prices[t]
        cash = 0.0
        rows.append({
            "date": dt.isoformat(),
            "total_value": total_value,
            "holdings_value": value_holdings,
            "cash": cash
        })
    df_res = pd.DataFrame(rows)
    return df_res

# ----------------- Exemplo de execução -----------------
if __name__ == "__main__":
    as_of = AS_OF
    print("Rodando seleção Bazin em", as_of)
    df_all, selection = select_universe(TICKERS, as_of)
    print("\nTodas as avaliações:")
    print(df_all[["ticker","cash_yield","dividend_yield","liquidity_brl","debt_to_ebitda","passes","reasons"]].sort_values("cash_yield", ascending=False).to_string(index=False))
    print("\nSeleção final (top):")
    print(selection[["ticker","cash_yield","dividend_yield","liquidity_brl","debt_to_ebitda"]].to_string(index=False))

    # Opcional: backtest simples com as ações selecionadas (se houver seleção)
    if not selection.empty:
        sel_tickers = selection["ticker"].tolist()
        start_bt = as_of - timedelta(days=365*3)  # backtest 3 anos
        end_bt = as_of
        print("\nExecutando backtest simples (exemplo) com rebalance semestral...")
        bt = simple_backtest(sel_tickers, start_bt, end_bt, initial_capital=10000.0, rebalance_months=6)
        print(bt.tail())
    else:
        print("Nenhuma ação passou nos filtros com os parâmetros atuais.")
```

```python
"""
Backtest Bazin com reinvestimento automático de proventos (DRIP)
Dependências: yfinance, pandas, numpy
pip install yfinance pandas numpy
"""

import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from typing import List

# ---------------- Configurações ----------------
TICKERS = ["VALE3.SA", "ITUB4.SA", "PETR4.SA"]   # selecione seus tickers (usar sufixo .SA para B3)
START_DATE = "2020-01-01"
END_DATE = datetime.today().strftime("%Y-%m-%d")
INITIAL_CAPITAL = 10000.0
REBALANCE_MONTHS = 6        # rebalance semestral (6) — troque para 12 = anual, 3 = trimestral...
ALLOW_FRACTIONAL_SHARES = True  # permite compras fracionadas ao reinvestir
# ------------------------------------------------

def fetch_price_and_dividends(tickers: List[str], start: str, end: str):
    """
    Busca histórico de preços (Close) e dividends via yfinance.
    Retorna:
      prices_df: DataFrame index=Date, columns=tickers contendo Close
      divs: dict[ticker] -> Series (date -> dividend_per_share)
    """
    prices = {}
    dividends = {}
    for t in tickers:
        tk = yf.Ticker(t)
        hist = tk.history(start=start, end=end, actions=False)  # includes trading days
        if hist.empty:
            prices[t] = pd.Series(dtype=float)
        else:
            prices[t] = hist["Close"].rename(t)
        divs = tk.dividends
        # filtrar intervalo
        if not divs.empty:
            divs = divs[(divs.index.date >= pd.to_datetime(start).date()) & (divs.index.date <= pd.to_datetime(end).date())]
        dividends[t] = divs
    # construir DataFrame de preços com índice de datas de mercado unificado
    all_idx = pd.Index(sorted(set().union(*[s.index for s in prices.values()])))
    prices_df = pd.DataFrame(index=all_idx)
    for t, s in prices.items():
        prices_df[t] = s.reindex(all_idx)
    return prices_df, dividends

def generate_rebalance_dates(start: pd.Timestamp, end: pd.Timestamp, months_interval: int):
    """
    Gera lista de datas de rebalance (usando o último dia útil do mês como referência),
    começando no primeiro dia disponível >= start.
    """
    dates = pd.date_range(start=start, end=end, freq='M').to_pydatetime().tolist()
    if not dates:
        return []
    rebalance = []
    for i, d in enumerate(dates):
        if i % months_interval == 0:
            rebalance.append(pd.Timestamp(d).date())
    return rebalance

def backtest_with_drip(selection_tickers: List[str],
                       start_date: str,
                       end_date: str,
                       initial_capital: float,
                       rebalance_months: int,
                       allow_fractional: bool = True):
    """
    Simulação com:
      - compra inicial equal-weight no primeiro dia de mercado >= start_date
      - reinvestimento automático (DRIP): quando há dividendo, comprar mais ações da mesma ação ao preço de fechamento do mesmo dia
      - rebalance equal-weight a cada rebalance_months meses (ajusta posições para igual valor)
    Retorna DataFrame diário com valor de portfólio e detalhe das posições em checkpoints (rebalance).
    """
    prices_df, dividends = fetch_price_and_dividends(selection_tickers, start_date, end_date)
    # remover colunas sem dados
    prices_df = prices_df.dropna(how='all', axis=1)
    tickers = prices_df.columns.tolist()
    if len(tickers) == 0:
        raise ValueError("Nenhum dado de preço disponível para os tickers selecionados.")
    # timeline (apenas dias de mercado disponíveis no prices_df)
    timeline = prices_df.index.normalize().unique()
    start_ts = pd.to_datetime(start_date)
    end_ts = pd.to_datetime(end_date)
    timeline = timeline[(timeline >= start_ts) & (timeline <= end_ts)]
    if len(timeline) == 0:
        raise ValueError("Nenhuma data de negociação encontrada no intervalo informado.")
    # rebalance dates
    rebalance_dates = generate_rebalance_dates(timeline[0], timeline[-1], rebalance_months)

    # inicializar posições
    holdings = {t: 0.0 for t in tickers}
    cash = 0.0

    # compra inicial equal-weight no primeiro dia de mercado
    first_date = timeline[0]
    prices_first = prices_df.loc[first_date]
    valid_prices = prices_first.dropna()
    active_tickers = valid_prices.index.tolist()
    n_active = len(active_tickers)
    alloc_per = initial_capital / n_active
    for t in active_tickers:
        p = prices_first[t]
        shares = alloc_per / p
        if not allow_fractional:
            shares = int(shares)
        holdings[t] = shares
    cash = initial_capital - sum(holdings[t] * prices_first[t] for t in active_tickers)

    # registros diários
    records = []
    # map de dividendos por ticker por data (datetime.date)
    div_map = {}
    for t in tickers:
        s = dividends.get(t)
        if s is None or s.empty:
            div_map[t] = {}
        else:
            # converter index para date
            div_map[t] = {d.date(): float(v) for d, v in s.items()}

    # iterar timeline
    for current in timeline:
        date = pd.Timestamp(current).date()
        prices_today = prices_df.loc[current]
        # 1) tratar dividendos ocorridos no dia: para cada ticker que pagou hoje,
        #    calcular montante = dividend_per_share * holdings[t]
        #    e reinvestir imediatamente comprando shares ao preço de fechamento do dia (se preço disponível)
        for t in tickers:
            if date in div_map.get(t, {}):
                div_per_share = div_map[t][date]
                if holdings.get(t, 0) > 0:
                    amount = div_per_share * holdings[t]  # valor recebido
                    p = prices_today.get(t, np.nan)
                    if np.isfinite(p) and p > 0:
                        # reinvestir na mesma ação
                        added_shares = amount / p
                        if not allow_fractional:
                            added_shares = int(added_shares)
                        holdings[t] += added_shares
                        # sem cash remanescente, pois reinvestimos tudo; se preferir, guarde em cash:
                        # cash += amount - added_shares * p
                    else:
                        # preço não disponível hoje -> acumular em cash (opcional)
                        cash += amount

        # 2) rebalance se hoje for data de rebalance (usa preço de fechamento de hoje)
        if date in rebalance_dates:
            # valor total do portfólio (holdings * price) + cash
            total_value = cash
            for t in tickers:
                p = prices_today.get(t, np.nan)
                if np.isfinite(p):
                    total_value += holdings.get(t, 0) * p
            # alocação equal-weight
            target_per = total_value / len(tickers)
            # ajustar holdings para cada ticker
            new_holdings = {}
            new_cash = 0.0
            for t in tickers:
                p = prices_today.get(t, np.nan)
                if not np.isfinite(p) or p <= 0:
                    new_holdings[t] = holdings.get(t, 0)
                    continue
                target_shares = target_per / p
                if not allow_fractional:
                    target_shares = int(target_shares)
                new_holdings[t] = target_shares
            # recomputar cash leftover (assumindo execução a preço de fechamento)
            used = sum(new_holdings[t] * prices_today[t] for t in tickers if np.isfinite(prices_today.get(t, np.nan)))
            new_cash = total_value - used
            holdings = new_holdings
            cash = new_cash

        # registrar valor do portfólio no final do dia
        total_value = cash
        detail = {}
        for t in tickers:
            p = prices_today.get(t, np.nan)
            if np.isfinite(p):
                total_value += holdings.get(t, 0) * p
            detail[t] = holdings.get(t, 0)
        records.append({
            "date": date.isoformat(),
            "total_value": total_value,
            "cash": cash,
            **{f"hold_{t}": detail[t] for t in tickers}
        })

    df_results = pd.DataFrame(records).set_index("date")
    return df_results

# ------------------ Exemplo de execução ------------------
if __name__ == "__main__":
    sel = TICKERS
    print("Rodando backtest Bazin com DRIP para:", sel)
    df_bt = backtest_with_drip(sel, START_DATE, END_DATE, INITIAL_CAPITAL, REBALANCE_MONTHS, ALLOW_FRACTIONAL_SHARES)
    print(df_bt.tail(10))
    # salvar resultados
    df_bt.to_csv("backtest_bazin_drip_results.csv")
    print("Backtest salvo em backtest_bazin_drip_results.csv")
```
